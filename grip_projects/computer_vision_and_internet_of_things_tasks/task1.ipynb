# Character detector which extracts printed or handwritten text from an image - task one

importing packages and libraries

```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 8, 16
```

'en' specifies english language

```python
reader = easyocr.Reader(['en'])
```

```python
Image("ai_overview.png")
```

```python
output = reader.readtext("ai_overview.png")
```

```python
output
```

```python
cord = output[-1][0]
```

```python
x_min, y_min = [int(min(idx)) for idx in zip(*cord)]
```

```python
x_max, y_max = [int(max(idx)) for idx in zip(*cord)]
```

```python
image = cv2.imread('ai_overview.png')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
